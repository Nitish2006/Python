```python
import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9, 11, 13]
y = [2, 4, 6, 8, 10, 12, 14]
m = [1,2,3,4,5,6,7]
plt.subplot(1,2,2)
plt.plot(x, y,color='purple',linestyle='--',marker='o')
plt.xlabel('price')
plt.ylabel('QD')
plt.title('Price vs Quantity Demanded')
plt.grid(True)
plt.subplot(1,2,1)
plt.plot(x,m)
plt.title("PVQD@")
plt.show()
```

Bar

```python
category=['a','b','v','c','d']
values=[5,7,9,11,13]
plt.bar(category,values,color='green')
plt.show()
```

Histogram


```python
val1=[10,20,25,35,45,67,87]
plt.hist(val1,bins=10,color='red',edgecolor='black')
plt.show()
```

Scatter

```python
plt.scatter(x,m,marker='o')
plt.show()
```

Bubble

```python
sizes=[10,50,100,150,200,250]
plt.scatter(x,y,sizes=sizes,marker='x')
plt.show()
```

Pie

```python
import matplotlib.pyplot as plt

labels = ['a', 'b', 'c', 'd']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellow', 'lightcoral', 'lightyellow']
explode = (0.2, 0, 0, 0)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, colors=colors, explode=explode)
plt.show()
```
